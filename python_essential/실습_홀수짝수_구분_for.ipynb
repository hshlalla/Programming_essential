### 실습_08.자연수의 홀수 짝수 구분
1. 1에서 20까지 3으로 나우어 떨어지는 요소를 갖는 num 리스트를 만들고
2. num의 숫자의 홀수 ,짝수를 판별하시요
3. 출력 예(3은 홀수 입니다)

```python
num = list(range(0,20,3))[1:]
num1=[]
num2=[]
for x in num:
    if (x%2)==0:
        
        num1.append(x)
    else:

        num2.append(x)
print("{}짝수입니다".format(num1))
print("{}홀수입니다".format(num2))
```

```python
num = list(range(0,20,3))  
for i in num[1:]:
    if (i%2==0):
        print("{}는 짝수 입니다".format(i))
    else:
        print("{}는 홀수 입니다".format(i))
```

```python
num = list(range(0,20,3))[1:]

for i in num:
    if (i%2==0):
        print("{}는 짝수 입니다".format(i))
    else:
        print("{}는 홀수 입니다".format(i))
```
