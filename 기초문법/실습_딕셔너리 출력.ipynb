####  (실습 12) 딕셔너리에서 value가 78인 key를 찾아서 출력 (key = ???? ) 
1. d = {'1':[60,80,30],'2':[78,54,98],'3' : [100,50,70],'4' : [70,60,80]}
2. 78점 점수를 가지고 있는 학생(key) 은 누가? 

```python
d= {"1":[60,80,30],"2":[78,54,98], "3":[100,50,70], "4":[70,60,80]}
score=80
for i in d.values():
    if i==score:
        print(i)
```

```python
# 데이터 찾기
for i in d.items():
    print(i)
    if ((i[1][0] ==78) or (i[1][1]==78) or (i[1][2]==78)):
        print("Key = ",i[0])
```
