## ( 실습_10.  for, if, range, type 혼합 )
1. 6개의 원소를 [1, 9, "one", "two", 3, "three"] 갖는 리스트 val을 만들고, 
2. 마지막 원소를 자기 이름(ex."jswoo")으로 바꿉니다. 
3. 그리고 -4번째 인덱스를 갖는 원소와 
4. 리스트의 첫번째 문자열("") 데이터를 출력하시요.

```python
val=[1, 9, "one", "two", 3, "three"]
```

```python
val[5]="jswoo";val
```

```python
val[-4]
```

```python
for i in val:
    if type(i)==str:
        print("-4번째 문장은 {}, 첫번째 문자열은 {}".format(val[-4],i))
        break
```

```python
for i in val:
    if type(i)==str:
        find_str = i
        break
print("-4번째 문장은 {}, 첫번째 문자열은 {}".format(val[-4],find_str))
```

```python
for i in val:
    if type(i) == str:
        break
print("-4번째 문자는 {}.첫번째 문자열은 {}".format(val[-4],i))
```

```python
for i in range(len(val)):
    if type(val[i]) == str:
        alpha_index = i
        break
print("-4번째 문자는 {}.첫번째 문자열은 {}".format(val[-4],val[alpha_index]))
```

```python
for i in enumerate(val):
    print(i)
```
