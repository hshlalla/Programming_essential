```python
import matplotlib.pyplot as plt
import seaborn as sns
iris = sns.load_dataset("iris")
iris.head(3)
iris.columns
```

```python
iris_index = iris["species"].value_counts().index
iris_values = iris["species"].value_counts().values
iris_index
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = iris_index
sizes = iris_values



plt.pie(sizes, #값
        labels=labels, #분류, 구분 class
        autopct="%1.1f%%", #갑표시
        explode=(0,0.1,0), #강조하고자 하는 값의 값을 띄워서 강조시킨다
        shadow=True,
        startangle=180)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
def chart_pie(x):
    tips_index = tips[x].value_counts().index
    tips_values = tips[x].value_counts().values
    labels = tips_index
    sizes = tips_values

    plt.pie(sizes, #값
        labels=labels,
        autopct="%1.1f%%",
        shadow=True,
        startangle=90)
    plt.show()
```
