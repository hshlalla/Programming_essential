### (실습.컬럼별 생존 유무에 따른 분포 함수 만들기)
1.stacked_barchart()
2.stacked_barchart("sex")

```python
def stacked_barchart(x):
  survived=titanic[titanic["survived"]==1][x].value_counts()
  dead=titanic[titanic["survived"]==0][x].value_counts()
  df=pd.DataFrame([survived,dead])
  df.plot(kind="bar",stacked=True, figsize=(6,4))
```

```python
stacked_barchart("pclass")
```

```python
stacked_barchart("embarked")
```

```python
stacked_barchart("sibsp")
```

```python
ti_cols=['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare',
       'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town',
       'alive', 'alone']
for qq in ti_cols:
    stacked_barchart(qq)
```
